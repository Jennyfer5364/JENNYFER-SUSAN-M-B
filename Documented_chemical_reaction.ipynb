The module chemistry is imported to support the chemical reaction.
To import chemistry module use the syntax 

> ***from chempy import chemistry***



```python
from chempy import chemistry
```

###**Reaction function to represent the chemical reaction**


1.   The elements used are:


> 1. Hydrogen and oxygen
> 2. Nitrogen and oxygen





```python
reaction1 = chemistry.Reaction({'N2': 2, 'O2': 1},
                              {'N2O': 2})
```

```python
reaction2 = chemistry.Reaction({'H2': 2, 'O2': 1},
                              {'H2O': 2})
```

**Print the reaction obtained**

```python
print(reaction1)
print(reaction2)
```

**Print the order of the reaction**

```python
print(reaction1.order())
print(reaction2.order())
```
